### 1. If statement on a subset

```python
#Creating price-range flags on subset
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
#Looking for max value in df
df['prices'].max()
```

### 2. Loc function on a subset

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### 4. Busiest day column - If-statements with for-loops

```python
#Checking frequency of orders by day of the week
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Creating a loop to find busyness of days
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Printing loop results
result
```

```python
#Creating a new column containing the loop results for respective rows
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### 6. Busiest hours of the day

#### Sorting: Most orders 9am to 16pm, Average orders 6am to 9am and 17pm to 12am and Fewest orders 1am to 5am

```python
#Creating a loop to find busyness of hours of a day

result2 = []

for value in ords_prods_merge["hour_of_day"]:
  if value == 9:
    result2.append("Most orders")
  elif value == 10:
    result2.append("Most orders")
  elif value == 11:
    result2.append("Most orders")
  elif value == 12:
    result2.append("Most orders")
  elif value == 13:
    result2.append("Most orders")
  elif value == 14:
    result2.append("Most orders")
  elif value == 15:
    result2.append("Most orders")
  elif value == 16:
    result2.append("Most orders")
  elif value == 6:
    result2.append("Average orders")
  elif value == 7:
    result2.append("Average orders")
  elif value == 8:
    result2.append("Average orders")
  elif value == 17:
    result2.append("Average orders")
  elif value == 18:
    result2.append("Average orders")
  elif value == 19:
    result2.append("Average orders")
  elif value == 20:
    result2.append("Average orders")
  elif value == 21:
    result2.append("Average orders")
  elif value == 22:
    result2.append("Average orders")
  elif value == 23:
    result2.append("Average orders")
  elif value == 0:
    result2.append("Average orders")
  else:
    result2.append("Fewest orders")
```

```python
result2
```

```python
#Creating a new column containing the loop results for respective rows counting two days
ords_prods_merge['busiest_period_of_day'] = result2
```
